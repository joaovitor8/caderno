```python
#  Escreva um programa para calcular quantos dias levará para a colônia de uma bactéria A ultrapassar ou igualar a colônia de uma bactéria B, com base nas taxas de crescimento de 3% e 1,5% respectivamente. Considere que a colônia A inicia com 4 elementos e a B com 10.

colonia_a = 4
colonia_b = 10
dias = 0

while colonia_a <= colonia_b:
  colonia_a = colonia_a * 3 / 100 + colonia_a
  colonia_b = colonia_b * 1.5 / 100 + colonia_b
  dias += 1

print(colonia_a)
print(colonia_b)
print(f'Levará {dias} dias para a colônia de A ultrapassar a de B.')
```

```python
# Para tratar uma quantidade de 15 dados de avaliações de pessoas usuárias de um serviço da empresa, precisamos verificar se as notas são válidas. Então, escreva um programa que vai receber a nota de 0 a 5 de todos os dados e verificar se é um valor válido. Caso seja inserido uma nota acima de 5 ou abaixo de 0, repita até que a pessoa usuária insira um valor válido.

for i in range(15):
  while True:
    try:
      nota = int(input(f"Digite a nota do usuário {i+1} (de 0 a 5): "))
      if 0 <= nota <= 5:
        break
      else:
        print("Nota inválida. Digite um valor entre 0 e 5.")
    except ValueError:
      print("Entrada inválida. Digite um número inteiro.")

  print(f"Nota do usuário {i+1}: {nota}")
```

```python
# Desenvolva um programa que leia um conjunto indeterminado de temperaturas em Celsius e informe a média delas. A leitura deve ser encerrada ao ser enviado o valor -273°C.

temperaturas = []

while True:
  temperatura = float(input("Digite a temperatura em Celsius (ou -273 para encerrar): "))

  temperaturas.append(temperatura)

  if temperatura == -273:
    break

temperaturas = temperaturas[:-1]

media = sum(temperaturas) / len(temperaturas)

print(f"A média das temperaturas é: {media:.2f}°C")
```

```python
# Escreva um programa que calcule o fatorial de um número inteiro fornecido pela pessoa usuária. Lembrando que o fatorial de um número inteiro é a multiplicação desse número por todos os seus antecessores até o número 1. Por exemplo, o fatorial de 5 é 5 x 4 x 3 x 2 x 1 = 120.


n = int(input("Digite um número inteiro: "))

fatorial = 1

for i in range(1, n + 1):
  fatorial *= i

print(f"O fatorial de {n} é: {fatorial}")
```

```python
# Os números primos possuem várias aplicações dentro da Ciência de Dados em criptografia e segurança, por exemplo. Um número primo é aquele que é divisível apenas por um e por ele mesmo. Assim, faça um programa que peça um número inteiro e determine se ele é ou não um número primo.

def is_prime(n):
  """Verifica se um número é primo."""
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

n = int(input("Digite um número inteiro: "))

if is_prime(n):
  print(f"O número {n} é primo.")
else:
  print(f"O número {n} não é primo.")
```

```python
# Vamos entender a distribuição de idades de pensionistas de uma empresa de previdência. Escreva um programa que leia as idades de uma quantidade não informada de clientes e mostre a distribuição em intervalos de [0-25], [26-50], [51-75] e [76-100]. Encerre a entrada de dados com um número negativo.

jovem = []
adulta = []
meia_idade = []
anciao = []

while True:
  idade = int(input("Digite a idade do cliente (ou um número negativo para encerrar): "))

  if idade <= 25:
    jovem.append(idade)
  elif idade > 25 and idade <= 50:
    adulta.append(idade)
  elif idade > 50 and idade <= 75:
    meia_idade.append(idade)
  elif idade > 75 and idade <= 100:
    anciao.append(idade)

  if idade < 0:
    break

print("Distribuição de idades:")
print(f"Jovem: {len(jovem)}")
print(f"Adulta: {len(adulta)}")
print(f"Meia-idade: {len(meia_idade)}")
print(f"Ancião: {len(anciao)}")
```

```python
# prompt:  Em uma eleição para gerência em uma empresa com 20 pessoas colaboradoras, existem quatro candidatos(as). Escreva um programa que calcule o(a) vencedor(a) da eleição. A votação ocorreu da seguinte maneira:
# Cada colaborador(a) votou em uma das quatro pessoas candidatas (que representamos pelos números 1, 2, 3 e 4).
# Também foram contabilizados os votos nulos (representados pelo número 5) e os votos em branco (representados pelo número 6).
# Ao final da votação, o programa deve exibir o número total de votos para cada candidato(a), os nulos e os votos em branco. Além disso, deve calcular e exibir a porcentagem de votos nulos em relação ao total de votos e a porcentagem de votos em branco em relação ao total de votos.

votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0
votos_candidato4 = 0
votos_nulos = 0
votos_brancos = 0

for i in range(20):
  voto = int(input(f"Digite o voto do colaborador {i+1} (1, 2, 3, 4, 5 - nulo, 6 - branco): "))
  if voto == 1:
    votos_candidato1 += 1
  elif voto == 2:
    votos_candidato2 += 1
  elif voto == 3:
    votos_candidato3 += 1
  elif voto == 4:
    votos_candidato4 += 1
  elif voto == 5:
    votos_nulos += 1
  elif voto == 6:
    votos_brancos += 1
  else:
    print("Voto inválido.")

print("\nResultados da Eleição:")
print("Votos para o Candidato 1:", votos_candidato1)
print("Votos para o Candidato 2:", votos_candidato2)
print("Votos para o Candidato 3:", votos_candidato3)
print("Votos para o Candidato 4:", votos_candidato4)
print("Votos Nulos:", votos_nulos)
print("Votos em Branco:", votos_brancos)

total_votos = votos_candidato1 + votos_candidato2 + votos_candidato3 + votos_candidato4 + votos_nulos + votos_brancos

if total_votos > 0:
  percentual_nulos = (votos_nulos / total_votos) * 100
  percentual_brancos = (votos_brancos / total_votos) * 100
  print(f"Percentual de Votos Nulos: {percentual_nulos:.2f}%")
  print(f"Percentual de Votos em Branco: {percentual_brancos:.2f}%")

# Encontrar o vencedor
vencedor = max(votos_candidato1, votos_candidato2, votos_candidato3, votos_candidato4)
if vencedor == votos_candidato1:
  print("\nO Candidato 1 venceu a eleição!")
elif vencedor == votos_candidato2:
  print("\nO Candidato 2 venceu a eleição!")
elif vencedor == votos_candidato3:
  print("\nO Candidato 3 venceu a eleição!")
elif vencedor == votos_candidato4:
  print("\nO Candidato 4 venceu a eleição!")
else:
  print("\nHouve empate entre os candidatos.")
```
