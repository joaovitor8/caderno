Calculadora com operadores

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

soma = n1 + n2

print(f'A soma dos números é: {soma}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))
n3 = int(input('Digite mais um número: '))

soma = n1 + n2 + n3

print(f'A soma dos números é: {soma}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

subtracao = n1 - n2

print(f'A subtracao dos números é: {subtracao}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

multiplicacao = n1 * n2

print(f'A multiplicação dos números é: {multiplicacao}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número, diferente de 0: '))

divisao = n1 / n2

print(f'A divisão dos números é: {divisao}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

exponenciacao = n1 ** n2

print(f'O resultado da exponenciação dos números é: {exponenciacao}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

divisao_inteira = n1 // n2

print(f'O resultado da divisão inteira dos números é: {divisao_inteira}')
```

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

resto = n1 % n2

print(f'O resto da divisão dos números é: {resto}')
```

```python
nota1 = float(input('Digite a nota 1: '))
nota2 = float(input('Digite a nota 2: '))
nota3 = float(input('Digite a nota 3: '))

media = (nota1 + nota2 + nota3) / 3

print(f'A média das notas é: {media}')
```

```python
n1 = 5
n2 = 12
n3 = 20
n4 = 15

peso1 = 1
peso2 = 2
peso3 = 3
peso4 = 4

media_ponderada = (n1 * peso1 + n2 * peso2 + n3 * peso3 + n4 * peso4) / (peso1 + peso2 + peso3 + peso4)

print(f'A média ponderada das notas é: {media_ponderada}')
```

```python
frase = input("Digite uma frase: ")
nova_frase = frase.replace("a", "@")
nova_frase
```
