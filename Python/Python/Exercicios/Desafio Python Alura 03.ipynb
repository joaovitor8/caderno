```python
def obter_informacoes_numero(numero):
  """
  Retorna informações sobre um número, incluindo se é par ou ímpar,
  positivo ou negativo e inteiro ou decimal.
  """
  if numero % 2 == 0:
    par_impar = "par"
  else:
    par_impar = "ímpar"

  if numero > 0:
    positivo_negativo = "positivo"
  elif numero < 0:
    positivo_negativo = "negativo"
  else:
    positivo_negativo = "neutro"

  if numero == int(numero):
    inteiro_decimal = "inteiro"
  else:
    inteiro_decimal = "decimal"

  return f"O número é {par_impar}, {positivo_negativo} e {inteiro_decimal}."


num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

print("Qual operação deseja realizar?")
print("1. Adição (+)")
print("2. Subtração (-)")
print("3. Multiplicação (*)")
print("4. Divisão (/)")

opcao = input("Digite o número da operação desejada: ")

if opcao == '1':
  resultado = num1 + num2
  print(f"O resultado da adição é: {resultado}")
  print(obter_informacoes_numero(resultado))
elif opcao == '2':
  resultado = num1 - num2
  print(f"O resultado da subtração é: {resultado}")
  print(obter_informacoes_numero(resultado))
elif opcao == '3':
  resultado = num1 * num2
  print(f"O resultado da multiplicação é: {resultado}")
  print(obter_informacoes_numero(resultado))
elif opcao == '4':
  if num2 == 0:
    print("Erro: Divisão por zero!")
  else:
    resultado = num1 / num2
    print(f"O resultado da divisão é: {resultado}")
    print(obter_informacoes_numero(resultado))
else:
  print("Opção inválida!")
```

```python
def verifica_triangulo(lado1, lado2, lado3):
  """Verifica se os lados formam um triângulo e classifica seu tipo."""

  if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    if lado1 == lado2 == lado3:
      return "Triângulo Equilátero"
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
      return "Triângulo Isósceles"
    else:
      return "Triângulo Escaleno"
  else:
    return "Não forma um triângulo"

lado1 = float(input("Digite o primeiro lado do triângulo: "))
lado2 = float(input("Digite o segundo lado do triângulo: "))
lado3 = float(input("Digite o terceiro lado do triângulo: "))

resultado = verifica_triangulo(lado1, lado2, lado3)
print(resultado)
```

```python
preco_etanol = 1.70
preco_diesel = 2.00

litros_vendidos = float(input("Digite a quantidade de litros vendidos: "))
tipo_combustivel = input("Digite o tipo de combustível (E para etanol, D para diesel): ").upper()

if tipo_combustivel == 'E':
  if litros_vendidos <= 15:
    desconto = 0.02
  else:
    desconto = 0.04
  valor_desconto = preco_etanol * litros_vendidos * desconto
  valor_a_pagar = (preco_etanol * litros_vendidos) - valor_desconto
elif tipo_combustivel == 'D':
  if litros_vendidos <= 15:
    desconto = 0.03
  else:
    desconto = 0.05
  valor_desconto = preco_diesel * litros_vendidos * desconto
  valor_a_pagar = (preco_diesel * litros_vendidos) - valor_desconto
else:
  print("Tipo de combustível inválido.")
  valor_a_pagar = 0

print(f"Valor a ser pago: R$ {valor_a_pagar:.2f}")
```

```python
vendas_2022 = float(input("Digite a quantidade de vendas em 2022: "))
vendas_2023 = float(input("Digite a quantidade de vendas em 2023: "))

variacao_percentual = ((vendas_2023 - vendas_2022) / vendas_2022) * 100

print(f"A variação percentual de vendas entre 2022 e 2023 foi de {variacao_percentual:.2f}%")

if variacao_percentual > 20:
  print("Sugestão: Bonificação para o time de vendas.")
elif variacao_percentual >= 2:
  print("Sugestão: Pequena bonificação para o time de vendas.")
elif variacao_percentual >= -10:
  print("Sugestão: Planejamento de políticas de incentivo às vendas.")
else:
  print("Sugestão: Corte de gastos.")
```
